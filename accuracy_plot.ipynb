```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

cv_df = pd.read_csv('model_fold_accuracy.csv')
cv_df.columns = ["Unnamed:","model", "fold_idx", "accuracy"]

# sns.boxplot(x='model', y='accuracy', data=cv_df[3:])
sns.stripplot(x='model', y='accuracy', data=cv_df[3:], 
              size=8, jitter=True, edgecolor="gray", linewidth=3)

labely = plt.ylabel("ACCURACY")
labelx = plt.xlabel("MODEL")
labely.set_color("black")
labelx.set_color("black")
plt.savefig('model_accuracy.pdf')
plt.show()
```
